##the columns "MCAL" & "DCAL" are highly correlated with depth so we removed the column

```python
data.drop(["MCAL","DCAL"],axis=1,inplace=True)
```

# principle component extraction of the features in which variance of the data are high

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=7)
```

```python
df=pca.fit_transform(data)
```

```python
df.shape
```

```python
from sklearn.cluster import KMeans
```

```python
int=[]
ren=[]
for i in range(1,10):
    km=KMeans(n_clusters=i)
    km.fit(df)
    int.append(km.inertia_)
    ren.append(i)
    
```

```python
df
```

```python
####geting number of cluster from elbow methods
```

```python
plt.plot(ren,int)
```

```python
kmeans=KMeans(n_clusters=3)

```

```python
kmeans.fit(df)
```

# prediction for cluster for each and every row

```python
pred=kmeans.predict(df)
```

```python
data["pred"]=pred
```

```python
data
```

```python
data
```

```python
plt.plot(data.Depth,data.pred)
```
